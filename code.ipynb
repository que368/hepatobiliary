# 数据处理

```python
# 读取编号与疾病对应
from openpyxl import load_workbook

wb=load_workbook(r"E:\Desktop\毕业论文\症状.xlsx")
sheets=wb.worksheets

sheet3=sheets[2]
print(sheet3)
```

```python
# 第一列为数字编号
col1=[]
for col in sheet3['A']:
    col1.append(col.value)
print(col1)
```

```python
# 第二列为中文编号
col2=[]
for col in sheet3['B']:
    col2.append(col.value)
print(col2)
```

```python
rows=sheet3.rows
print(rows.value)
print(sheet3['A1'])
```

```python
# 编号与疾病中文的对应字典
zhengzhuang=dict(zip(col1,col2))
print(zhengzhuang)
```

```python
# 输出该处理的数据
sheet4=sheets[3]
s4col1=[]
for col in sheet4['A']:
    s4col1.append(col.value)
print(s4col1)
```

```python
# 数字编号替换为文字
list=[]
# print(zhengzhuang.keys())
for n in s4col1:
#     print(n)
    if n in zhengzhuang.keys():
        list.append(zhengzhuang[n])
print(list)
```

```python
# 写入工作簿
import xlsxwriter
workbook=xlsxwriter.Workbook(r"E:\Desktop\毕业论文\症状.xlsx")
worksheet=workbook.add_worksheet('sheet5')
worksheet.write_column('D2',list)
workbook.close()
```

```python
# 构建知识图谱
with open('E:/Desktop/毕业论文/数据/!肝胆病(症).csv','r',encoding='gbk') as f:
    reader =csv.reader(f)
    for item in reader:
        if (reader.line_num==1):
            continue
        end_node1=Node("证",name=item[2])
        end_node2=Node("病",name=item[1],id=item[0])
        start_node=Node("症",name=item[3])
        g.merge(end_node2,"病","name")
        g.merge(start_node,"症","name")
        relation_z_b=Relationship(start_node,item[5],end_node2)//症状疾病
        g.create(relation_z_b)
        if item[2] is "":
            continue    
        g.merge(end_node1,"证","name")
        print("当前行数：",reader.line_num,"当前内容：",item)
        relation_z_z=Relationship(start_node,item[4],end_node1)//症状证候
        relation_zh_b=Relationship(end_node2,item[6],end_node1)//证候疾病
        g.create(relation_z_z)
        g.create(relation_zh_b)
```

```python
# 删除知识图谱全部节点
g.run('match (n) detach delete n')
```
